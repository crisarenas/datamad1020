#### Use the appropriate visulalization to show the distribution of the column `Age`

```python
# your code here
'''Yo creo que una buena forma es el histograma. 
Permite ver que la mayoria de personas tienen entorno a 28 años.'''
age = titanic['Age']
plt.hist(age,color='pink')

plt.title('Age distribution')

plt.show()
```

#### Use subplots to plot the distribution of the `Age` variable with bins equal to `10`, `20` and `50`.

```python
# your code here
plt.figure(figsize=(15,4))
plt.subplot(1,3,1) 
plt.title("Bins = 10")
plt.hist(age,bins = 10, color='pink')

plt.subplot(1,3,2)
plt.title("Bins = 20")
plt.hist(age,bins = 20, color='pink')

plt.subplot(1,3,3)
plt.title("Bins = 30")
plt.hist(age,bins = 30, color='pink')

plt.show()
```

#### Whether the bin size affects your plot? Why?


Aumenta el número de barritas, lo que nos permite ser más exactos. Si al principio deciamos que la mayoría de persona 
tenía entorno a 28 años, ahora vemos que ese número se acerca más a 20.
