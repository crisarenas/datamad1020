```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
import seaborn as sn
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
```

```python
from sklearn import datasets

data = datasets.load_wine()

X = pd.DataFrame(data["data"], columns=data["feature_names"])
y = pd.Series(data["target"])
```

```python
print(data.DESCR)
```

## 1. Train a KMeans clustering model on the data set using 8 clusters and compute the silhouette score for the model.

##### Info de google:
The silhouette score falls within the range [-1, 1].

The silhouette score of 1 means that the clusters are very dense and nicely separated. The score of 0 means that clusters are overlapping. The score of less than 0 means that data belonging to clusters may be wrong/incorrect.

```python
km = KMeans(n_clusters = 8).fit(X)
score = silhouette_score(X, km.labels_, metric='euclidean')

print(f'Silhouetter Score: {score}')
```

## 2. Train a KMeans clustering model on the data set using 5 clusters and compute the silhouette score for the model.

```python
km = KMeans(n_clusters = 5).fit(X)
score = silhouette_score(X, km.labels_, metric='euclidean')

print(f'Silhouetter Score: {score}')
```

## 3. Train a KMeans clustering model on the data set using 3 clusters and compute the silhouette score for the model.

```python
km = KMeans(n_clusters = 3).fit(X)
score = silhouette_score(X, km.labels_, metric='euclidean')

print(f'Silhouetter Score: {score}')
```

## 4. Use elbow curve visualizations to see if you can determine the best number of clusters to use.

The Yellowbrick library has 3 metrics that you can plot using the `metric` parameter:

- **distortion**: mean sum of squared distances to centers
- **silhouette**: mean ratio of intra-cluster and nearest-cluster distance
- **calinski_harabaz**: ratio of within to between cluster dispersion

```python
models = [KMeans(i) for i in range(1,11)]
scores = [m.fit(X).score(X) for m in models]
print(scores)
plt.plot(range(1,11),scores);
```

```python
from yellowbrick.cluster import KElbowVisualizer


# Instantiate the clustering model and visualizer
model = KMeans()
visualizer = KElbowVisualizer(model, k=(2,10))

visualizer.fit(X)        # Fit the data to the visualizer
visualizer.show();        # Finalize and render the figure
```

**4 es el mejor numero de clusters**

## 5. Try performing the same elbow tests with an AgglomerativeClustering model and compare the results you get to the KMeans results.

```python
from sklearn.cluster import AgglomerativeClustering

# Instantiate the clustering model and visualizer
aglo = AgglomerativeClustering()
visualizer = KElbowVisualizer(aglo, k=(2,10))

visualizer.fit(X)        # Fit the data to the visualizer
visualizer.show();        # Finalize and render the figure
```

Tambien coincide en que lo mejor son 4 clusters, aunque la verdad es que con 3 clusters el score de Kmeans era un pelin mayor

```python
km = KMeans(n_clusters = 4).fit(X)
score = silhouette_score(X, km.labels_, metric='euclidean')

print(f'Silhouetter Score: {score}')
```

## 7. Apply a PCA transform and plot the first two principle components with the plot point colors determined by cluster.

```python
aglo3 = AgglomerativeClustering(n_clusters = 3).fit(X_scale)
X["labels_aglo3"] = aglo3.labels_


pca = PCA(2)
data_t = pca.fit_transform(X[["alcohol","hue"]])
plt.grid()
sn.scatterplot(data_t[:,0],data_t[:,1], hue= X["labels_aglo3"]);


```

## 8. Generate a series of t-SNE plots showing the clusters at a variety of perplexities.

```python
fig = plt.figure(figsize=(15,15))
index=0
for p in [3,4,5,8,10,20]:
    index += 1
    plt.subplot(3, 3, index)
    plt.title(f"Perplexity = {p}")
    data = TSNE(n_components = 2, perplexity = p).fit_transform(X)
    sn.scatterplot(x = data[:,0], y = data[:,1], hue= X["labels_aglo3"]);
```
